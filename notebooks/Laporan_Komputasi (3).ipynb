##**1.1 Latar Belakang**

Perkembangan teknologi digital telah mendorong pemanfaatan pengolahan sinyal dalam berbagai bidang ilmu pengetahuan dan teknologi. Salah satu bentuk sinyal yang banyak digunakan adalah sinyal akustik atau sinyal audio. Sinyal akustik merupakan representasi gelombang suara yang direkam dan disimpan dalam bentuk data digital sehingga dapat dianalisis menggunakan metode komputasi. Pemrosesan sinyal audio banyak diterapkan pada sistem komunikasi, pengenalan suara, multimedia, instrumentasi, hingga sistem monitoring yang memanfaatkan informasi suara sebagai sumber data utama. Pemanfaatan teknologi digital pada sinyal audio memungkinkan proses analisis dilakukan secara lebih cepat, akurat, dan efisien dibandingkan metode konvensional (Supardi et al., 2022).

Dalam proses akuisisi data, sinyal audio yang diperoleh dari mikrofon atau perangkat perekam umumnya tidak hanya mengandung informasi utama yang diinginkan, tetapi juga mengandung gangguan berupa noise. Noise merupakan sinyal yang tidak diinginkan dan dapat berasal dari lingkungan sekitar, gangguan elektronik pada perangkat, maupun proses transmisi data. Keberadaan noise menyebabkan kualitas sinyal menurun sehingga informasi yang terkandung di dalamnya menjadi lebih sulit untuk dianalisis. Oleh karena itu, diperlukan suatu metode yang mampu memisahkan komponen sinyal utama dari komponen noise agar kualitas data yang diperoleh dapat ditingkatkan (Putri et al., 2025).

Salah satu pendekatan yang banyak digunakan dalam pengolahan sinyal digital adalah analisis frekuensi menggunakan Transformasi Fourier. Transformasi Fourier memungkinkan sinyal yang semula direpresentasikan dalam domain waktu diubah menjadi domain frekuensi sehingga frekuensi-frekuensi penyusun sinyal dapat diamati secara lebih jelas. Dalam implementasi komputasi modern, proses transformasi ini umumnya dilakukan menggunakan algoritma Fast Fourier Transform (FFT). FFT merupakan pengembangan dari Discrete Fourier Transform (DFT) yang mampu melakukan perhitungan spektrum frekuensi secara lebih cepat dan efisien sehingga sangat sesuai digunakan untuk analisis sinyal digital dalam jumlah besar. Penggunaan FFT memungkinkan identifikasi frekuensi dominan pada sinyal audio serta memudahkan proses analisis karakteristik frekuensi yang terkandung di dalamnya (Supardi et al., 2022).

Setelah spektrum frekuensi diperoleh, proses pengurangan noise dapat dilakukan melalui teknik filtering. Salah satu metode filtering yang umum digunakan adalah low-pass filter. Low-pass filter bekerja dengan mempertahankan komponen frekuensi rendah yang mengandung informasi utama sinyal dan meredam komponen frekuensi tinggi yang sering kali diasosiasikan sebagai noise. Melalui proses ini, kualitas sinyal dapat ditingkatkan tanpa mengubah karakteristik utama sinyal secara signifikan. Penelitian yang dilakukan oleh Putri et al. (2025) menunjukkan bahwa penerapan low-pass filter efektif dalam mereduksi noise pada sinyal audio dan mampu meningkatkan kualitas hasil pemrosesan sinyal berdasarkan analisis respon frekuensi dan kualitas audio yang dihasilkan.

Berdasarkan latar belakang tersebut, pada studi kasus ini dilakukan analisis sinyal akustik yang telah terkontaminasi noise menggunakan metode Fast Fourier Transform (FFT). Sinyal terlebih dahulu dianalisis pada domain frekuensi untuk mengidentifikasi komponen frekuensi dominan dan komponen noise. Selanjutnya diterapkan low-pass filter untuk menghilangkan frekuensi yang tidak diinginkan, kemudian sinyal direkonstruksi kembali ke domain waktu menggunakan Inverse Fast Fourier Transform (IFFT). Melalui tahapan tersebut diharapkan dapat dipahami bagaimana metode FFT dan filtering frekuensi dapat digunakan untuk meningkatkan kualitas sinyal akustik yang mengalami gangguan noise.

##**3.1 Pembacaan dan Visualisasi Sinyal Akustik**

Pada studi kasus ini digunakan data sinyal akustik berupa file audio berformat WAV yang telah terkontaminasi oleh noise. Data dibaca menggunakan fungsi wavfile.read() dari pustaka SciPy untuk memperoleh informasi dasar sinyal berupa sampling rate dan jumlah sampel. Hasil pembacaan menunjukkan bahwa sinyal memiliki sampling rate sebesar 44.100 Hz dengan jumlah sampel sebanyak 132.300 sampel, sehingga diperoleh durasi sinyal sekitar 3 detik.

**CODE :**

```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

fs, y = wavfile.read("../data/noisy_signal.wav")

if len(y.shape) > 1:
    y = y[:,0]

# Jumlah sampel
N = len(y)

# Array waktu
t = np.arange(N) / fs

print("Sampling Rate =", fs, "Hz")
print("Jumlah Sampel =", N)

# Plot 0.1 detik pertama
durasi_plot = 0.1

mask = t <= durasi_plot

plt.figure(figsize=(12,5))
plt.plot(t[mask], y[mask])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Raw Noisy Acoustic Signal")
plt.show()
```

| Jumlah Sampel | 132.300 |
| Durasi Sinyal | 3 s |

Berdasarkan Gambar 3.1, sinyal akustik menunjukkan pola osilasi yang kompleks dengan amplitudo yang berubah-ubah terhadap waktu. ##**3.2 Transformasi Sinyal ke Domain Frekuensi Menggunakan FFT**

Pada tahap ini, sinyal akustik yang telah direpresentasikan dalam domain waktu pada subbab sebelumnya ditransformasikan ke domain frekuensi menggunakan Fast Fourier Transform (FFT). Tujuan dari proses ini adalah untuk menguraikan sinyal menjadi komponen frekuensi penyusunnya sehingga karakteristik spektralnya dapat dianalisis secara lebih jelas.

**CODE :**

```python
Y = np.fft.fft(y)

freqs = np.fft.fftfreq(N, d=1/fs)

magnitude = np.abs(Y)

plt.figure(figsize=(12,5))

plt.plot(freqs[:N//2], magnitude[:N//2])

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency Spectrum of the Acoustic Signal")

plt.xlim(0, 5000)

plt.grid(True)
plt.show()
```

FFT mengubah sinyal dari domain waktu x(t) menjadi representasi frekuensi X(f) sehingga setiap komponen frekuensi dapat dilihat kontribusinya terhadap sinyal total.

Berdasarkan Gambar 3.2, terlihat bahwa sinyal memiliki beberapa puncak pada frekuensi tertentu yang menunjukkan keberadaan frekuensi dominan dalam sinyal. Distribusi spektrum yang tidak merata menunjukkan bahwa sinyal tidak hanya terdiri dari satu frekuensi tunggal, melainkan merupakan hasil superposisi beberapa komponen frekuensi. Setelah spektrum frekuensi diperoleh melalui FFT, dilakukan identifikasi terhadap komponen frekuensi yang menyusun sinyal. Tahap ini bertujuan untuk membedakan frekuensi dominan yang merepresentasikan sinyal utama dengan frekuensi lain yang diasumsikan sebagai noise.

###**Gambar 3.3** Spektrum Frekuensi Sinyal Akustik


Berdasarkan Gambar 3.3, terlihat bahwa sebagian besar energi sinyal terkonsentrasi pada rentang frekuensi rendah yang ditunjukkan oleh puncak spektrum dengan magnitudo tinggi. Untuk memisahkan sinyal utama dari noise, ditentukan nilai cutoff frequency sebesar 900 Hz. Pemilihan nilai ini dilakukan berdasarkan inspeksi visual terhadap spektrum FFT, di mana frekuensi dominan masih berada di bawah batas tersebut. Dengan demikian, frekuensi di bawah 900 Hz dipertahankan sebagai sinyal utama, sedangkan frekuensi di atasnya akan direduksi pada proses filtering. Penentuan cutoff frequency menjadi langkah penting karena memengaruhi kualitas sinyal hasil rekonstruksi. Nilai cutoff yang terlalu rendah berpotensi menghilangkan informasi penting dari sinyal, sedangkan nilai yang terlalu tinggi dapat menyebabkan sebagian noise tetap lolos ke tahap berikutnya.

##**3.4 Penerapan Low-Pass Filter pada Domain Frekuensi**

Setelah nilai cutoff frequency ditentukan, dilakukan proses filtering pada domain frekuensi menggunakan metode low-pass filter. Filter ini bekerja dengan mempertahankan komponen frekuensi yang berada di bawah nilai cutoff dan menghilangkan komponen frekuensi yang berada di atasnya. Pada penelitian ini digunakan nilai cutoff frequency sebesar 900 Hz, yang diperoleh berdasarkan hasil identifikasi spektrum frekuensi pada subbab sebelumnya.

**CODE :**

```python
f_cutoff = 900

Y_filtered = Y.copy()

Y_filtered[np.abs(freqs) > f_cutoff] = 0

plt.figure(figsize=(12,5))
plt.plot(freqs[:N//2], np.abs(Y_filtered[:N//2]))

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Filtered Frequency Spectrum")
plt.xlim(0,1500)

plt.grid(True)
plt.show()
```

###**Gambar 3.4** Spektrum Frekuensi Setelah Penerapan Low-Pass Filter

Berdasarkan Gambar 3.4, terlihat bahwa komponen frekuensi di atas 900 Hz telah berhasil direduksi sehingga spektrum frekuensi menjadi lebih sederhana dibandingkan sebelum filtering. Puncak-puncak frekuensi dominan pada rentang frekuensi rendah masih tetap dipertahankan, sedangkan komponen frekuensi tinggi yang sebelumnya tersebar pada spektrum telah berkurang secara signifikan.

Hasil tersebut menunjukkan bahwa low-pass filter mampu memisahkan sebagian besar noise dari sinyal utama tanpa menghilangkan karakteristik frekuensi yang penting. Dengan berkurangnya komponen frekuensi tinggi yang tidak diinginkan, sinyal yang diperoleh menjadi lebih representatif terhadap informasi utama yang terkandung dalam data akustik. Spektrum hasil filtering ini kemudian digunakan sebagai masukan pada proses rekonstruksi sinyal menggunakan Inverse Fast Fourier Transform (IFFT).


##**3.5 Rekonstruksi Sinyal Menggunakan Inverse FFT**

Setelah proses low-pass filtering selesai dilakukan pada domain frekuensi, sinyal direkonstruksi kembali ke domain waktu menggunakan Inverse Fast Fourier Transform (IFFT). Proses ini bertujuan untuk memperoleh bentuk sinyal hasil filtering sehingga dapat dibandingkan dengan sinyal asli yang masih mengandung noise.

**CODE :**

```python
y_clean = np.fft.ifft(Y_filtered)

y_clean = np.real(y_clean)

mask = t <= 0.1

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(t[mask], y[mask])
plt.title("Original Noisy Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(t[mask], y_clean[mask])
plt.title("Filtered Signal (After IFFT)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```

###**Gambar 3.5** Perbandingan Sinyal Asli dan Sinyal Hasil Rekonstruksi

Berdasarkan Gambar 3.5, terlihat perbedaan antara sinyal asli dan sinyal hasil rekonstruksi setelah proses filtering. Sinyal hasil rekonstruksi memiliki bentuk gelombang yang lebih halus dengan fluktuasi amplitudo berfrekuensi tinggi yang lebih sedikit dibandingkan sinyal asli. Hal ini menunjukkan bahwa sebagian besar komponen noise telah berhasil direduksi melalui proses filtering pada domain frekuensi.

Meskipun terjadi pengurangan noise, pola utama sinyal tetap dapat dipertahankan. Kondisi ini menunjukkan bahwa pemilihan cutoff frequency yang digunakan mampu menghilangkan komponen frekuensi yang tidak diinginkan tanpa menghilangkan informasi penting yang terkandung dalam sinyal. Dengan demikian, proses rekonstruksi menggunakan IFFT berhasil menghasilkan sinyal yang lebih bersih dan lebih representatif terhadap sinyal asli sebelum terkontaminasi noise.


##**3.6 Analisis Efektivitas Filtering**

Hasil analisis menunjukkan bahwa metode Fast Fourier Transform (FFT) dan low-pass filter mampu mengurangi komponen noise pada sinyal akustik secara efektif. Hal ini ditunjukkan oleh perubahan spektrum frekuensi setelah proses filtering, di mana komponen frekuensi tinggi yang diasumsikan sebagai noise berhasil direduksi tanpa menghilangkan frekuensi dominan yang merepresentasikan sinyal utama.

Perbandingan antara sinyal sebelum dan sesudah rekonstruksi menggunakan IFFT juga menunjukkan bahwa sinyal hasil filtering memiliki bentuk gelombang yang lebih halus dan stabil. Fluktuasi amplitudo berfrekuensi tinggi yang sebelumnya muncul akibat noise mengalami penurunan, sehingga kualitas sinyal menjadi lebih baik dan lebih mudah untuk dianalisis.

Keunggulan filtering pada domain frekuensi dibandingkan domain waktu terletak pada kemampuannya untuk menargetkan rentang frekuensi tertentu secara langsung. Dengan FFT, komponen sinyal dan noise dapat dipisahkan berdasarkan karakteristik frekuensinya sehingga proses reduksi noise dapat dilakukan secara lebih selektif. Pendekatan ini lebih efektif dibandingkan metode perataan (smoothing) pada domain waktu yang berpotensi mengubah bentuk sinyal utama.

Secara keseluruhan, kombinasi FFT, low-pass filtering, dan IFFT berhasil meningkatkan kualitas sinyal akustik yang terkontaminasi noise. Metode ini memungkinkan pemulihan informasi utama yang terkandung dalam sinyal sekaligus mengurangi pengaruh gangguan yang tidak diinginkan, sehingga hasil analisis menjadi lebih representatif terhadap kondisi sinyal yang sebenarnya.
